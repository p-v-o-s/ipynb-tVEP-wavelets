```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATASET_NAME = "2017-08-09-SN001_CWV_R-HC-O2_V5_epochs"
V_epochs = np.loadtxt("data/%s.csv" % DATASET_NAME, delimiter=",")
V_SCALE = 1e-6
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
for V in V_epochs:
    ax1.plot(V/V_SCALE,'-', alpha=0.1)
    
ax1.plot(V_epochs.mean(axis=0)/V_SCALE,'-', color="k", label="mean VEP")
ax1.legend()
ax1.set_xlabel("Time (ms)")
ax1.set_ylabel("EEG ($\mu$V)")
ax1.set_title(DATASET_NAME)
```
